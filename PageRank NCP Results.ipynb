```python
from typing import *
import localgraphclustering as lgc
import numpy as np

import matplotlib.pyplot as plt
import functools

import sys
import os
sys.path.insert(0, os.path.join("..", "LocalGraphClustering", "notebooks"))
import helper

NTHREADS=192
NITER=1000000000
DEEP=True

def run_apr_ncp(g, gname, nthreads=NTHREADS, deep=DEEP, 
                spectral_args={'iterations':NITER}, **kwargs):
    ratio = 1.0
    if g._num_vertices > 1000000:
        ratio = 0.05
    elif g._num_vertices > 100000:
        ratio = 0.1
    elif g._num_vertices > 10000:
        ratio = 0.4
    elif g._num_vertices > 7500:
        ratio = 0.6
    elif g._num_vertices > 5000:
        ratio = 0.8
    print(gname, "ratio =", ratio)
    ncp = lgc.NCPData(g)
    ncp.approxPageRank(ratio=ratio, nthreads=nthreads, deep=deep, 
                       spectral_args=spectral_args, **kwargs)
    fig = lgc.NCPPlots(ncp).cond_by_size()[0]
    fig.axes[0].set_title(gname + " Conductance by Size (Deep Search)")
    fig.savefig("figures/deep-apr-ncp-"+gname+".pdf", bbox_inches="tight", figsize=(3,3))
    # remove the results of deep searches
    df = ncp.as_data_frame()
    df_no_deep = df[~df["method"].str.contains("deep")]
    fig = lgc.NCPPlots(df_no_deep).cond_by_size()[0]
    fig.axes[0].set_title(gname + " Conductance by Size")
    fig.savefig("figures/apr-ncp-"+gname+".pdf", bbox_inches="tight", figsize=(3,3))
    return fig, ncp

for (gname,gfile) in {'netscience':'netscience-cc.edges'}.items():
    g = lgc.GraphLocal(os.path.join("..", "data", gfile),
                                 'edgelist', ' ')
    g.discard_weights()
    fig, ncp = run_apr_ncp(g, gname)    
```
